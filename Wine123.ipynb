```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv('wine.csv')
```

```python
X=wine.iloc[:,1:]
X.head()
```

```python
Y=wine.iloc[:,0]
Y.head()
```

```python
#data Normalization
from sklearn import preprocessing
X=preprocessing.StandardScaler().fit_transform(X)
```

```python
X[0:5]
```

```python
#Train Test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
Y_test.shape
Y
```

```python
#Training and Predicting
from sklearn.neighbors import KNeighborsClassifier
knmodel=KNeighborsClassifier(n_neighbors=5)
knmodel.fit(X_train,Y_train)
ypredict=knmodel.predict(X_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
acc=accuracy_score(Y_test,ypredict)
acc
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(Y_test,ypredict)
cm
```

```python
cm1=pd.DataFrame(data=cm,index=['1','2','3'],columns=['1','2','3'])
cm1
```

```python
#Output Visualazation
prediction_op=pd.DataFrame(data=[Y_test.values,ypredict],index=["Y_test","ypredict"])
```

```python
prediction_op.transpose()
```

```python
#Finding the K Value
ks=21
mean_acc=np.zeros((ks-1))

for i in range(1,ks):
    neigh=KNeighborsClassifier(n_neighbors=i).fit(X_train,Y_train)
    yhat=neigh.predict(X_test)
    mean_acc[i-1]=accuracy_score(Y_test,yhat)
    
```

```python
print(mean_acc)
```

```python
plt.plot(range(1,ks),mean_acc,'g')
plt.ylabel("Accuracy")
plt.xlabel("K")
plt.tight_layout()
plt.show()
```
